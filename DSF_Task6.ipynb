# Introduction to NumPy and Pandas for Data Manipulation

## NumPy Array Creation

```python
import numpy as np
arr1 = np.array([1,2,3,4,5])
print('1D Array:', arr1)

arr2 = np.array([[1,2,3],[4,5,6]])
print('2D Array:\n', arr2)

print('Addition:', arr1 + 5)
print('Multiplication:', arr1 * 2)
print('Slicing:', arr1[1:4])
```

## Array Statistics

```python
print('Mean:', np.mean(arr1))
print('Median:', np.median(arr1))
print('Standard Deviation:', np.std(arr1))
print('Index of Max:', np.argmax(arr1))
```

## Pandas Series & DataFrame

```python
import pandas as pd

s = pd.Series([10,20,30], index=['a','b','c'])
print('Series:\n', s)

data = {'Name':['Ali','Sara','John'], 'Age':[20,21,19]}
df = pd.DataFrame(data)
print('DataFrame:\n', df)
```

## Data Importing

```python
df_csv = pd.read_csv('cleaned_dataset.csv') 
print(df_csv.head())
print(df_csv.tail())
```

## Data Cleaning

```python
df_csv.fillna(0, inplace=True)
df_csv.dropna(inplace=True)
print(df_csv)
```

## Indexing & Selection

```python
print(df.loc[0])  # By label
print(df.iloc[1])  # By index
```

## Sorting & Filtering

```python
print(df.sort_values(by='Age'))
print(df[df['Age'] > 19])
```

## GroupBy & Aggregations

```python
print(df.groupby('Age').mean(numeric_only=True))
```

## Merging & Joining

```python
df2 = pd.DataFrame({'Name':['Ali','Sara'], 'Score':[90,85]})
merged = pd.merge(df, df2, on='Name')
print(merged)

concat = pd.concat([df, df2], axis=0, ignore_index=True)
print(concat)
```

## Descriptive Statistics

```python
print(df.describe())
```

## Line Plot Basics

```python
import matplotlib.pyplot as plt

values = [1,2,3,4,5]
plt.plot(values)
plt.title('Line Plot')
plt.show()
```

## Bar Chart Frequencies

```python
categories = ['A','B','C']
counts = [5,7,3]
plt.bar(categories, counts)
plt.title('Bar Chart')
plt.show()
```

## Scatter Plot Relationships

```python
x = [5,7,8,7,2,17]
y = [99,86,87,88,100,86]
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.show()
```

## Seaborn Pairplot

```python
import seaborn as sns

iris = sns.load_dataset('iris')
sns.pairplot(iris)
plt.show()
```

## Category Box Plot

```python
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.show()
```

## Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_iris = iris.select_dtypes(include=['number'])

sns.heatmap(numeric_iris.corr(), annot=True, cmap='coolwarm')
plt.show()
```

## Subplot Grid

```python
fig, axs = plt.subplots(2,2)
axs[0,0].plot([1,2,3])
axs[0,1].bar([1,2,3],[3,2,1])
axs[1,0].scatter([1,2,3],[3,2,1])
axs[1,1].hist([1,2,3,1,2,3])
plt.show()
```

## Seaborn Styling

```python
sns.set_style('darkgrid')
sns.histplot(iris['sepal_length'], kde=True, color='green')
plt.title('Styled Plot')
plt.xlabel('Sepal Length')
plt.show()
```

## Violin Plot Distribution

```python
sns.violinplot(x='species', y='petal_length', data=iris)
plt.show()
```

## Hybrid Visualization

```python
sns.histplot(iris['sepal_length'], kde=True)
plt.title('Hybrid KDE + Histogram')
plt.show()
```
